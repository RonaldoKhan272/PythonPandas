# Nilai Kumulatif pada Pandas Data Frame

# Import Modules

```python
import pandas as pd

print(pd.__version__)
```

# Persiapan Data Frame

Kolom pemain dan kolom gol yang di cetak

```python
d={'pemain':['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', 'Joni'],
   'goal':[2, 1, 3, 1, 1, 2, 2, 3]}

df = pd.DataFrame(d)
df
```

# Menampilkan nilai kumulatif (cummulative)

Nilai kumulatif gol = `cumsum()` atau sum
, Cara Baca :
- index nol merupakan jumlah gol dari index nol ditambah index satu
- begitu selanjutnya

```python
df['goal'].cumsum().to_frame()
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
```

Dikategorikan berdasarkan pemain dengan menggunakan fungsi `groupby()`

```python
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

Kumulatif maximum dengan fungsi `cummax()`

```python
df['cummax'] = df['goal'].cummax()
df
```

Kumulatif minimum dengan fungsi `cummin()`

```python
df['cummin'] = df['goal'].cummin()
df
```

Kumulatif product dengan fungsi `cumprod()` -> Hasil total perkalian

```python
df['cumprod'] = df['goal'].cumprod()
df
```
